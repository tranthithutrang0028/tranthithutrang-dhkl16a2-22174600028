```python
import pandas as pd

path_drinks = '/Users/nguyenxuanty/Documents/baitapchuong4/drinks.csv'

drinks = pd.read_csv(path_drinks)
drinks
```

```python
# kiểu dữ liệu (type) và kích thước (shape) của drink
print('kiểu dữ liệu của drinks là:', drinks.dtypes)
print('kích thước của drinks là:',drinks.shape)
```

```python
# cho biết số lượng bia tiêu thụ trung bình ở mỗi châu lục
mean_beer_serving = drinks.groupby('continent')
result_mean_beer = mean_beer_serving.aggregate({'beer_servings' : 'mean'})
print('số lượng bia tiêu thụ trung bình ở mỗi châu lục:\n',result_mean_beer )
```

```python
# thông tin thống kê tổng quát
print('thống kê tổng quát của drinks :\n',drinks.describe)
```

```python
# số lượng tiêu thụ rượi vang ở mỗi châu lục
mean_wine_serving = drinks.groupby('continent')
result_mean_wine = mean_wine_serving.aggregate({'wine_servings' : 'mean'})
print('số lượng rượu vang tiêu thụ trung bình ở mỗi châu lục:\n',result_mean_wine )
```
